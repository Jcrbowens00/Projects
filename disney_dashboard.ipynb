# Historic Disney Stock Price Dashboard (Feb 2023)

```python
import yfinance as yf
import mplfinance as mpf
```

```python
ticker = "DIS"
```

Disney Stock Over 4 years (2019 - 2023)

```python
history2 = yf.Ticker(ticker).history(period="48mo")
```

```python
mpf.plot(history2,type='line')
```

Disney Stock Price over the Past year (2022 - 2023)

```python
history = yf.Ticker(ticker).history(period="12mo")
```

```python
history.tail()
```

```python
mpf.plot(history, type='line', figratio=(18,10), mav=(7))
```

Disney Stock Price and Value within the last 3 months (Nov 2022 - Feb 2023)

```python
history1 = yf.Ticker(ticker).history(period="3mo")
```

```python
mpf.plot(history1,type='candle',mav=(3,6,9),volume=True)
```

```python
print (f"3 month low: {int(history1.Low.min())} US Dollars")
print (f"3 month high: {int(history1.High.max())} US Dollars")
```
